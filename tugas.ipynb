```python
import pandas as pd

# Panggil/baca datanya
df_data_sampah = pd.read_csv('data_sampah-2021.csv')[['nama_kabupaten_kota', 'jumlah_produksi_sampah', 'tahun']]

# 1. Menghitung total produksi sampah di seluruh kabupaten/kota di Jawa Barat untuk tahun tertentu
tahuns = int(input("Masukkan tahun yang diinginkan: "))
total_sampahnya = 0
tahunnya_ada = False
for i, row in df_data_sampah.iterrows():
    if row['tahun'] == tahuns:
        total_sampahnya += row['jumlah_produksi_sampah']
        tahunnya_ada = True

if tahunnya_ada:
    print(f"Total produksi sampah di Jawa Barat pada tahun {tahuns}: {total_sampahnya} ton per hari")
else:
    print(f"Tidak ada data sampah untuk tahun {tahuns}")

# Ekspor hasil nomor 1 ke CSV
df_total_tahun_tertentu = pd.DataFrame({
    'tahun': [tahuns],
    'total_produksi_sampah': [total_sampahnya if tahunnya_ada else 0]
})
df_total_tahun_tertentu.to_csv('total_sampah_tahun_tertentu.csv', index=False)

# 2. Menghitung total produksi sampah setiap tahunnya
total_per_tahun = {}
for i, row in df_data_sampah.iterrows():
    tahun = row['tahun']
    total_per_tahun[tahun] = total_per_tahun.get(tahun, 0) + row['jumlah_produksi_sampah']

print("\nTotal produksi sampah setiap tahun:")
total_per_tahun_list = []
for tahun, total in total_per_tahun.items():
    print(f"{tahun}: {total} ton per hari")
    total_per_tahun_list.append({'tahun': tahun, 'total_produksi_sampah': total})

# Ekspor hasil nomor 2 ke CSV
df_total_per_tahun = pd.DataFrame(total_per_tahun_list)
df_total_per_tahun.to_csv('total_produksi_sampah_tahun.csv', index=False)

# 3. Menghitung total produksi sampah berdasarkan daerah (kabupaten/kota)
total_per_wilayah = {}
for i, row in df_data_sampah.iterrows():
    wilayah = row['nama_kabupaten_kota']
    total_per_wilayah[wilayah] = total_per_wilayah.get(wilayah, 0) + row['jumlah_produksi_sampah']

print("\nTotal produksi sampah berdasarkan daerah:")
total_per_wilayah_list = []
for wilayah, total in total_per_wilayah.items():
    print(f"{wilayah}: {total} ton per hari")
    total_per_wilayah_list.append({'nama_kabupaten_kota': wilayah, 'total_produksi_sampah': total})

# Ekspor hasil nomor 3 ke CSV
df_total_per_wilayah = pd.DataFrame(total_per_wilayah_list)
df_total_per_wilayah.to_csv('total_produksi_sampah_wilayah.csv', index=False)
```
